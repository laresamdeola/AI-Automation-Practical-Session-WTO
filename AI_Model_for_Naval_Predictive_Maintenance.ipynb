Linear Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
lin_regressor = LinearRegression()
lin_regressor.fit(X_train, y_train)
```

```python
y_pred = lin_regressor.predict(X_test)
```

```python
r2_score(y_pred, y_test)
```

```python
print(lin_regressor.coef_)
print(lin_regressor.intercept_)
```

Save the Models

```python
joblib.dump(lin_regressor, 'linear_model.pkl')
joblib.dump(tree_regressor, 'decision_tree_model.pkl')
```

```python
def naval_maintenace(new_data, model):
  '''
     new_data will be a numpy array of the following format:
     [[0, 1, 2, 3...amount of independent features]]

     model will be either 'linear' or 'tree', does not default
     to any, you'll have to state the model explicitly.
  '''
  # load the saved ML models
  loaded_linear_model = joblib.load('linear_model.pkl')
  loaded_tree_model = joblib.load('decision_tree_model.pkl')

  if model == 'linear':
    return loaded_linear_model.predict(new_data)
  elif model == 'tree':
    return loaded_tree_model.predict(new_data)
```
